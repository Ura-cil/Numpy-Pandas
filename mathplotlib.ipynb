```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
plt.xlabel("Sugar Level")
plt.ylabel("Number Of Patients")
plt.title("Blood Sugar Chart")

blood_sugar_men = [113, 85, 90, 150, 149, 88, 93, 115, 135, 80, 77, 82, 129]
blood_sugar_women = [67, 98, 89, 120, 133, 150, 84, 69, 89, 79, 120, 112, 100]

men_freq = [0,0,0]
wmen_freq = [0,0,0]

for e in blood_sugar_men:
    if e >= 80 and e<100:
        men_freq[0] += 1
    elif e >= 100 and e<125:
        men_freq[1] += 1
    elif e >= 125 and e<=150:
        men_freq[2] += 1

for e in blood_sugar_women:
    if e >= 80 and e<100:
        wmen_freq[0] += 1
    elif e >= 100 and e<125:
        wmen_freq[1] += 1
    elif e >= 125 and e<=150:
        wmen_freq[2] += 1

plt.hist([blood_sugar_men,blood_sugar_women], bins=[80,100,125,150], rwidth=0.95, color=['green','orange'],label=['men','women'])
plt.plot([85,105,130], men_freq,color = "red",label = "men freq")
plt.plot([95,120,145], wmen_freq,color = "black", label = "women freq")
plt.legend()
plt.savefig("blood_sugar_chart.jpg",bbox_inches="tight", pad_inches=1, transparent=True)

```
